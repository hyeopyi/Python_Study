# 서울시 교통사고 시각화

```python
import pandas as pd
```

```python
traffic_Seoul = pd.read_csv("../Data/newSeoul_2005_2019.csv")
traffic_Seoul.head()
```

```python
traffic_Seoul.info()
```

```python
# 2019년 데이터만 추출하기
traffic_Seoul_2019 = traffic_Seoul[traffic_Seoul['년도']==2019]
traffic_Seoul_2019
```

```python
# index 정리하기
traffic_Seoul_2019.reset_index(drop=True, inplace=True)
traffic_Seoul_2019
```

```python
import numpy as np
```

```python
# 자치구별 발생건수, 사망자수, 부상자수 집계하기
traffic_analysis = pd.pivot_table(traffic_Seoul_2019[['자치구명','발생건수','부상자수','사망자수']],
                                  index='자치구명',aggfunc=np.sum)
traffic_analysis.head()
```

```python
# 위도, 경도 데이터 불러오기
seoul_limit = pd.read_csv("../Data/seoul.csv", encoding='euc-kr')
seoul_limit.columns = ['자치구명', 'lon', 'lat']
seoul_limit.head()
```

```python
# traffic_seoul_2019와 seoul_limit 병합하기
data_result = pd.merge(traffic_analysis, seoul_limit, on='자치구명')
data_result.tail()
```

```python
# folium을 사용하기 위해 int를 float로 변환해야 한다
data_result[['발생건수','부상자수','사망자수']] = data_result[['발생건수','부상자수','사망자수']].astype(float)
data_result.info()
```

### 자치구의 구청을 표시하기


```python
import folium
```

```python
map = folium.Map(
    location=[37.5502,126.982],
    zoom_start=11)

# for name, lat, lon in zip(data_result.자치구명, data_result.lat, data_result.lon):
#         folium.Marker([lat, lon], popup=name).add_to(map)  

for n in data_result.index:
    folium.Marker([data_result['lat'][n], data_result['lon'][n]],
                popup=data_result['자치구명'][n]).add_to(map)

map
```

```python
# 자치구별 교통사고 발생건수를 표시하기
map = folium.Map(
    location=[37.5502,126.982],
    zoom_start=11)

for n in data_result.index:
    folium.Marker([data_result['lat'][n], data_result['lon'][n]],
                popup=data_result['자치구명'][n]).add_to(map)
    
    folium.CircleMarker([data_result['lat'][n], data_result['lon'][n]],
                        popup=data_result['발생건수'][n],
                        radius= data_result['발생건수'][n] /100,
                        fill_color = "skyblue",
                        fill_opacity = 0.3,
                        ).add_to(map)

map
```

```python
# 자치구별 교통사고 사망건수를 표시하기
map = folium.Map(
    location=[37.5502,126.982],
    zoom_start=11)

for n in data_result.index:
    folium.Marker([data_result['lat'][n], data_result['lon'][n]],
                popup=data_result['자치구명'][n]).add_to(map)
    
    folium.CircleMarker([data_result['lat'][n], data_result['lon'][n]],
                        popup=data_result['사망자수'][n],
                        radius= data_result['사망자수'][n] * 2,
                        color = 'red',
                        fill_color = "red",
                        fill_opacity = 0.2,
                        ).add_to(map)
 
map    
```

```python
# 자치구별 교통사고 사망건수를 최대, 최소 표시하기
map = folium.Map(
    location=[37.5502,126.982],
    zoom_start=11)

for n in data_result.index:
    folium.Marker([data_result['lat'][n], data_result['lon'][n]],
                popup=data_result['자치구명'][n]).add_to(map)
    

max_idx = data_result['사망자수'].argmax()
min_idx = data_result['사망자수'].argmin()

folium.CircleMarker([data_result['lat'][max_idx], data_result['lon'][max_idx]],
                        popup=data_result['사망자수'][max_idx],
                        radius= data_result['사망자수'][max_idx] * 2,
                        color = 'red',
                        fill_color = "red",
                        fill_opacity = 0.2,
                        ).add_to(map)

folium.CircleMarker([data_result['lat'][min_idx], data_result['lon'][min_idx]],
                        popup=data_result['사망자수'][min_idx],
                        radius= data_result['사망자수'][min_idx] * 2,
                        color = 'red',
                        fill_color = "red",
                        fill_opacity = 0.2,
                        ).add_to(map)

map    
```
