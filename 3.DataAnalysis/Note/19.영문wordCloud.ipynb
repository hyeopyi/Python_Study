```python
import nltk
import pickle
from nltk.corpus import stopwords
import re
nltk.download('all')
```

### text를 불러서 word 분류

```python
content = open('../Data/Independence.txt').read()
content
```

```python
# 불필요한 심벌 없애기
#cleaned_content = content.replace('!', '').replace(",", '')
cleanded_content = re.sub(r'[^\.\?\!\w\d\s]', '', content) # 문장 단위로 끊기
cleanded_content
```

```python
# 대문자를 소문자로 변환
cleanded_content = cleanded_content.lower()
cleanded_content
```

```python
# 문장에서 워드를 토큰으로 쪼갠다.
word_tokens = nltk.word_tokenize(cleanded_content)
print(word_tokens)
```

```python
# 품사분리
token_pos = nltk.pos_tag(word_tokens)
print(token_pos)
```

```python
# 명사만 추출하기
NN_words = []
for word, pos in token_pos:
    if 'NN' in pos:
        NN_words.append(word)
print(NN_words)
```

```python
# 원형으로 복원
wlem = nltk.WordNetLemmatizer()
lenmatized_words = []
for word in NN_words:
    new_word = wlem.lemmatize(word)
    lenmatized_words.append(new_word)
print(lenmatized_words)
```

```python
# 불용어 제거
stopwords_list = stopwords.words('english')
#stopwords_list
unique_NN_words = set(lenmatized_words) # 데이터 형태에따라 set 적용 - 중복값제거
final_NN_words = lenmatized_words

for word in unique_NN_words:
    if word in stopwords_list:
        while word in final_NN_words:
            final_NN_words.remove(word) # 불용어 한번에 제거
```

```python
# 원하는 단어 제거
customized_stopwords = ['be', 'today', 'yesterady', "it's", "don't", 'others']
unique_NN_words1 = set(final_NN_words)
for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 빈도분석
from collections import Counter

c = Counter(final_NN_words)  # 중요한 것은 count 결과, wordcolud에서는 count값을 주지 않는다, 분석을 위해 생성
print(c)
```

```python
print(c.most_common(20))
```

```python
# 워드 클라우드 
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud

# 워드클라우는 딕셔너리 등을 모른다 하나의 문장으로 만들어서 전달해야 한다
noun_text = ''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word
noun_text
```

```python
wordcloud = WordCloud(
    max_font_size=60,
    relative_scaling=0.5,
).generate(noun_text) # generate가 noun_text의 string value을 입력 받아 count 함.
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
