# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df[df['전출지별'].isnull()]
```

```python
df[df['전입지별'].isnull()]
```

```python
# NaN을 전(Prev) 데이터로 채움
df = df.fillna(method='ffill')  # 근처 값으로 채움
df.head()
```

```python
df.info()
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
df_seoul = df[mask]
df_seoul
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop(['전출지별'], axis=1, inplace=True)
df_seoul.head()
```

```python
# 전입지별 -> 전입지로 컬럼 이름 변경
df_seoul.rename({'전입지별':'전입지'},axis=1, inplace=True)
df_seoul
```

```python
# df_seoul의 index를 전입지로 변경
df_seoul.set_index('전입지', inplace=True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

```python
plt.plot(sr_one) #  값이 두개밖에 없어 sr_one 만드로 가능
plt.show()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title("서울 -> 경기 인구 이동")

# 축 이름 변경
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 그래프 꾸기미
# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14, 5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')

# y축 눈금 라벨 회전하기
plt.yticks(rotation = 45)

plt.title('서울 -> 경기 인구 이동')
plt.xlabel('이동 인구수')
plt.plot(sr_one)

# 범례 표시
plt.legend(labels=['서울->경기'], loc='best')
plt.show()
```

```python
# 그래프 꾸기미

# Style 서식 지정
plt.style.use('ggplot')

# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14, 7))

# x축 눈금 라벨 회전하기
plt.xticks(rotation='vertical')

# y축 눈금 라벨 회전하기
plt.yticks(rotation = 45)

plt.title('서울 -> 경기 인구 이동')
plt.xlabel('기간',size=15)
plt.ylabel('이동 인구수', size=15)

# Maker표시 
plt.plot(sr_one, marker="o", markersize=10)

# y축 범위 지정
plt.ylim(50000, 800000) # 최소값, 최대값

# 화살표 표시
plt.annotate('',#'화살표',
            xy=(20, 620000), # 화살표의 머리부분
            xytext=(3, 290000), # 화살표의 꼬리부분
            #xycoords='data', <- 안써도 됨
            arrowprops=dict(arrowstyle='->', color='skyblue', lw=5)
            )
            
plt.annotate('인구 이동 증가(1970~1995)', # 텍스트
             xy = (10, 400000),  # 텍스트 위치 기준점
             rotation = 31,     # 텍스트 회전각도
             va = 'baseline',    # 텍스트 상하정렬
             ha = 'center',     # 텍스트 좌우정렬
             fontsize = 12
            )      

plt.annotate('',
            xy=(45, 450000), # 화살표의 머리부분
            xytext=(29, 600000), # 화살표의 꼬리부분
            #xycoords='data', <- 안써도 됨
            arrowprops=dict(arrowstyle='->', color='green', lw=5)
            )

plt.annotate('인구 이동 감소(1995~2017)', # 텍스트
             xy = (38, 500000),  # 텍스트 위치 기준점
             rotation = -17,     # 텍스트 회전각도
             va = 'baseline',    # 텍스트 상하정렬
             ha = 'center',     # 텍스트 좌우정렬
             fontsize = 12
            )  

# 범례 표시
plt.legend(labels=['서울->경기'], loc='best', fontsize= 15)

# 이미지 저장
#plt.savefig("../Data/seoul.png")
plt.show()
```

### 동일한 그림에 여러개의 그래프를 추가


```python
# 서울에서 '충청남도', '경상북도', '강원도'로 이동한 인구데이터 값만 선택
df_3 = df_seoul.loc[['충청남도', '경상북도', '강원도']]
df_3
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
ax.plot(df_3.columns, df_3.loc['충청남도'],
        marker='o',
        markerfacecolor='green',
        color='olive',
        linewidth=2,
        label='서울->충남')

ax.plot(df_3.columns, df_3.loc['경상북도'],
        marker='o',
        markerfacecolor='blue',
        color='skyblue',
        linewidth=2,
        label='서울->경북')

ax.plot(df_3.columns, df_3.loc['강원도'],
        marker='o',
        markerfacecolor='red',
        color='magenta',
        linewidth=2,
        label='서울->강원')

ax.legend(loc='best')
ax.set_title('서울->충남, 경북, 강원 인구 이동', size=20)

ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)

ax.set_xticklabels(df_3.columns, rotation=90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x', labelsize=7)
ax.tick_params(axis='y', labelsize=7)
plt.show()
```

```python
# 서울에서 '충청남도', '경상북도', '강원도', '전라남도'로 이동한 인구데이터 값만 선택
df_4 = df_seoul.loc[['충청남도', '경상북도', '강원도', '전라남도']]
df_4
```

```python
# Style 서식 지정
plt.style.use('ggplot')

# 그래프 객채 생성(figure에 2개의 서브플롯을 생성)
fig = plt.figure(figsize=(25,15))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)


ax1.plot(df_4.columns, df_4.loc['충청남도'], marker='o', color='green', markerfacecolor='green', label='서울->충남')
ax2.plot(df_4.columns, df_4.loc['경상북도'], marker='o', color='skyblue', markerfacecolor='blue', label='서울->경북')
ax3.plot(df_4.columns, df_4.loc['강원도'], marker='o', color='magenta', markerfacecolor='red', label='서울->강원도')
ax4.plot(df_4.columns, df_4.loc['전라남도'], marker='o', color='yellow', markerfacecolor='orange', label='서울->전남')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

ax1.set_ylim(5000, 70000)
ax2.set_ylim(5000, 70000)
ax3.set_ylim(5000, 70000)
ax4.set_ylim(5000, 70000)

ax1.set_xticklabels(df_4.columns, rotation=75)
ax2.set_xticklabels(df_4.columns, rotation=75)
ax3.set_xticklabels(df_4.columns, rotation=75)
ax4.set_xticklabels(df_4.columns, rotation=75)

ax1.set_title('서울->충남 인구 이동')
ax2.set_title('서울->경북 인구 이동')
ax3.set_title('서울->강원 인구 이동')
ax4.set_title('서울->전남 인구 이동')


plt.show()
```

### 2010~2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
df_total = df_4[['합계']].sort_values(by='합계', ascending=True)
df_total
```
