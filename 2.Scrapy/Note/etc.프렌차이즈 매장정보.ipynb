```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
```

# 배스킨라빈스 매장명 수집

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
driver.get('http://www.baskinrobbins.co.kr/')
```

```python
driver.save_screenshot("../Data/001.png") # 화면을 관리하고 있는지 여부를 확인하기 위해 수행
```

```python
# 전국 정보 가져오기
#//*[@id="nform"]/fieldset/p[1]/span[1]/select

city_list_raw = driver.find_element(By.XPATH, '//*[@id="nform"]/fieldset/p[1]/span[1]/select')  # '' 사용
#print(gu_list_raw)

city_list = city_list_raw.find_elements(By.TAG_NAME, 'option')
#print(gu_list)

city_names= [option.get_attribute('value') for option in city_list]
city_names.remove('')
city_names
```

```python
# 검색버튼 click
# //*[@id="nform"]/fieldset/div[2]/button
xpath = '//*[@id="nform"]/fieldset/div[2]/button'
element_sle_city = driver.find_element(By.XPATH, xpath).click()
```

```python
# 한 지역 가져오기
element = driver.find_element(By.XPATH, '//*[@id="nform"]/fieldset/p[1]/span[1]/select')
element.send_keys(city_names[0])
xpath = '//*[@id="nform"]/fieldset/div[2]/button'
element_city_store = driver.find_element(By.XPATH, xpath).click()
```

```python
# BeautifulSoup으로 분석하기
res = driver.page_source
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
# #content > div > div.store_searcstore = soup.select("h3 > a")h > div.list > div > ul > li:nth-child(18) > article > h3 > a

names = soup.select("h3 > a")
# 매장명
for i in names:
    print(i.string)
```

```python
#content > div > div.store_search > div.list > div > ul > li:nth-child(1) > article > a > address > span

# 주소
list = soup.select("address > span")
for i in list:
    print(i.string)
```

```python
# 전지역 검색
storeName = []
storeAddress = []
import time
for city in city_names:

    element = driver.find_element(By.XPATH, '//*[@id="nform"]/fieldset/p[1]/span[1]/select')
    element.send_keys(city)
    time.sleep(3)

    
    xpath = '//*[@id="nform"]/fieldset/div[2]/button'
    element_sle_city = driver.find_element(By.XPATH, xpath).click() # 도시별 검색

    time.sleep(4)
    
    res = driver.page_source
    soup = BeautifulSoup(res, "html.parser")

    address = soup.select("address > span")
    for i in address:
        storeAddress.append(i.string)
    
    names = soup.select("h3 > a")
    for j in names:
        storeName.append(j.string)   

    time.sleep(2)
```

```python
# countLists만으로 DataFrame 생성
df = pd.DataFrame(data=storeName, columns=['매장명'])
df['주소'] = pd.Series(storeAddress)
df
```

```python
df.to_csv("../Data/Baskinrobbins.csv", encoding="utf-8", index=False)
```
