# 매개변수를 추가해 요청하는 방법
- 기상청 RSS
- 기상청 RSS는 URL에 지역번호를 지정하면 해당 지역 정보만 제공

지역정보(stnId)
- 전국 : 108
- 서울/경기도 : 109
- 강원도 : 105
- 충청북도 : 131
- 충청남도 : 133
- 전라북도 : 146
- 전라남도 : 156
- 경상북도 : 143
- 경상남도 : 159
- 제주특별자치도 : 184

```python
import urllib.request
import urllib.parse

# 기상청 RSS 사이트 주소
API = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"

# 매겨변수를 URL로 인코딩
values = {'stdId' : '109'}
params = urllib.parse.urlencode(values)

# URL 생성
url = API + "?" + params
print("url=", url)

# 다운로드
data = urllib.request.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```
